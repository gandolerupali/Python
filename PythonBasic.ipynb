**In Python, a token is the smallest unit in the source code. When you write a Python program, the code is broken down into individual tokens during the lexical analysis phase, also known as tokenization. Tokens are the building blocks of the language, and they represent the different elements in your code, such as keywords, identifiers, literals, operators, and punctuation.**

1.	Keywords
2.	Identifiers
3.	Literals

```python
# 14.	Take two numbers as input and find their greatest common divisor (GCD).

# Take input from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Use Euclidean algorithm
while num2 != 0:
    num1, num2 = num2, num1 % num2

# Display the result
print(f"The Greatest Common Divisor (GCD) is: {num1}")
```

```python
# Take input from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Find GCD using math module
import math
gcd = math.gcd(num1, num2)

# Display the result
print(f"The Greatest Common Divisor (GCD) of {num1} and {num2} is: {gcd}")
```

```python
# 20.	Write a Python program to check whether a number is divisible by both 3 and 5.

# Take a input
Num = int(input("Enter a number: "))

# Check whether a number is divisible by both 3 and 5

if Num % 3 == 0 and Num % 5 == 0:
  print(f"{Num} is divisible by both 3 and 5")
elif Num % 3 == 0 and Num % 5 != 0:
   print(f"{Num} is divisible by 3 but not by 5")
elif Num % 3 != 0 and Num % 5 == 0:
   print(f"{Num} is divisible by 5 but not by 3")
else:
  print(f"{Num} is not divisible by both 3 and 5")
```
